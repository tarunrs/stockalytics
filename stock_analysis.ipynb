```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("raw/INFY.csv")
```

```python
data
```

```python
temp = list(data["Close"].copy())
temp = [0] + temp[:-1]
len(temp)
data["prev_close"] = temp
```

```python
data["change"] = data["Close"] - data["prev_close"]
data["percentage_change"] = data["change"] * 100 / data["Close"]
```

```python
np.mean(data["percentage_change"])
```

```python
data = data.drop(data.index[0])
```
